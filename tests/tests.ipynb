```python
from docling.document_converter import DocumentConverter
```

```python
converter = DocumentConverter()
```

```python
source = "../data/Délibération SAN-2023-003 du 16 mars 2023 - Légifrance.pdf"  # PDF path or URL
result = converter.convert(source)
```

```python
result.document.save_as_markdown(filename="../data/result.md")
```

```python
# Get total number of pages
total_pages = len(result.document.pages)

# Extract markdown for each page
pages_markdown = [ result.document.export_to_markdown(page_no=i) for i in range(total_pages) ]
```
